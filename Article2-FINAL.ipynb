```python
#Imputation : Replace Missing Values:  ****** float by mean and Object by mode********
for column_name in df.columns:
    if df[column_name].dtype == 'float64' :
        df[column_name].fillna(df[column_name].mean(), inplace = True)
    else :
        df[column_name].fillna(df[column_name].mode()[0], inplace = True)
```

```python
#required importation for algorithms:
from sklearn import svm, tree, naive_bayes, ensemble , neighbors

#Common Model Helpers
from sklearn import feature_selection
from sklearn import model_selection
from sklearn import metrics

from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import VarianceThreshold

#Import SelectKBest for Feature Selection
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import mutual_info_classif

from sklearn.feature_selection import RFE
from sklearn.feature_selection import RFECV
```

```python
#list of the necessary functions to apply algorithms:
MLA = [
    ensemble.RandomForestClassifier(random_state=0),
    #Navies Bayes
    naive_bayes.BernoulliNB(),
    naive_bayes.GaussianNB(),
    #Trees    
    tree.DecisionTreeClassifier(random_state=0),
    tree.ExtraTreeClassifier(random_state=0),
    #Nearest Neighbor
    neighbors.KNeighborsClassifier(),
    #SVM
    svm.LinearSVC(random_state=0), 
    svm.NuSVC(probability=True),
    svm.SVC(probability=True) 
    ]
MLA2=[
     svm.LinearSVC(random_state=0), 
     ensemble.RandomForestClassifier(random_state=0),
     tree.DecisionTreeClassifier(random_state=0),
     tree.ExtraTreeClassifier(random_state=0),
] 


```

```python
#perapation of x and y:
ind_col = [col for col in df.columns if col != "'class'"]
X = ind_col
Y ="'class'"
x = df[ind_col]
y =df["'class'"]
y.shape
x.shape
```

```python
#splitting data into data test and data train:
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=model_selection.train_test_split(df[X], df[Y], test_size = 0.25, random_state = 0)
print(X_train.shape)
print(X_test.shape)
```

```python
#list of names of algorithms and its accuracy: 
MLA_columns = ['MLA Name', 'MLA Test Accuracy' ]
MLA_compare = pd.DataFrame(columns = MLA_columns)


row_index = 0
for alg in MLA:
    alg.fit(X_train, y_train)

    alg_acc = accuracy_score(y_test, alg.predict(X_test))
    MLA_name = alg.__class__.__name__
    MLA_compare.loc[row_index, 'MLA Name'] = MLA_name
    
    MLA_compare.loc[row_index, 'MLA Test Accuracy'] = alg_acc

   
    
    row_index+=1

    
MLA_compare.sort_values(by = ['MLA Test Accuracy'], ascending = False, inplace = True)
MLA_compare
```

```python
#importing necessary libraries for models:
from sklearn import svm, tree, naive_bayes, ensemble , neighbors
#Common Model Helpers
from sklearn import feature_selection
from sklearn import model_selection
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import VarianceThreshold
from sklearn.feature_selection import RFE
from sklearn.feature_selection import RFECV
```

```python
#number of selected features with different models using RFE:
MLA_columns = ['MLA Name' ,'number of feature selected','feature selected']
MLA_compare = pd.DataFrame(columns = MLA_columns)


row_index = 0
for alg in MLA2:
    
    selector = RFE(alg, step=1,  )
    selector.fit(x,y)
    new_b=[]
    
    number_of_features_selected=0
    s=selector.get_support()
    features_selected=[]
    
    for i in range(len(s)):
        if s[i]==True :
            features_selected.append(X[i])
            number_of_features_selected+=1
                
    
    MLA_name = alg.__class__.__name__
    MLA_compare.loc[row_index, 'MLA Name'] = MLA_name
    
    #MLA_compare.loc[row_index, 'MLA grid score'] = str(selector.grid_scores_)
    MLA_compare.loc[row_index, 'number of feature selected'] = str(number_of_features_selected)
    MLA_compare.loc[row_index, 'feature selected'] = str(features_selected)
    
    
    row_index+=1
  
    MLA_compare.sort_values(by = ['MLA Name'], ascending = False, inplace = True)
MLA_compare
```
